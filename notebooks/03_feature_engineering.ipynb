```python
# importing libraries and dataset 
import pandas as pd

df = pd.read_csv("../data/processed/cleaned_data.csv")
```

```python
#Feature 1: Addiction Risk (Target Variable)
#Convert Addiction Score into categories.
def addiction_risk(score):
    if score <= 4:
        return "Low"
    elif score <= 7:
        return "Medium"
    else:
        return "High"

df["Addiction_Risk"] = df["Addicted_Score"].apply(addiction_risk)
```

```python
df["Addiction_Risk"].value_counts()
```

```python
# Feature 2: Usage-Sleep Ratio
# Based on EDA, higher usage + lower sleep = higher addiction.
df["Usage_Sleep_Ratio"] = (
    df["Avg_Daily_Usage_Hours"]
    / df["Sleep_Hours_Per_Night"]
)
```

```python
# Feature 3: Wellbeing Score
# Combines sleep and mental health.
df["Wellbeing_Score"] = (
    df["Sleep_Hours_Per_Night"]
    + df["Mental_Health_Score"]
)
```

```python
# Feature 4: Social Conflict Intensity
# Since conflicts showed strong correlation.
df["Conflict_Intensity"] = pd.cut(
    df["Conflicts_Over_Social_Media"],
    bins=[0,3,6,10],
    labels=["Low","Medium","High"]
)
```

```python
# Feature 5: Academic Risk Flag
df["Academic_Risk"] = df[
    "Affects_Academic_Performance"
].map({
    "Yes":1,
    "No":0
})
```
